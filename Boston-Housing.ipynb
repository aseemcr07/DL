```python
import pandas as pd
```

```python
df = pd.read_csv("boston_housing.csv")
```

```python
df.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
data_Col = df.columns
pred = df[data_Col[data_Col!="medv"]]
targ = df["medv"]
```

```python
pred
```

```python
targ
```

```python
from keras.models import Sequential
from keras.layers import Dense
```

```python
model = Sequential()
```

```python
model.add(Dense(128, activation='relu', input_shape=(13, ), name='dense-1'))
model.add(Dense(64, activation='relu', name='dense-2'))
model.add(Dense(1, activation='linear', name='dense-output'))
```

```python
model.compile(optimizer='adam', loss='mse', metrics=['mae'])
```

```python
model.summary()
```

```python
#Splitting data
x_train, x_test, y_train, y_test = train_test_split(pred, targ, test_size=0.25, random_state=123, shuffle=True)
```

```python
# Training the model
model.fit(x_train, y_train, epochs=200, validation_split=0.2)
```

```python
import sklearn.metrics as skm
```

```python
predicted = model.predict(x_test)
```

```python
print("MAE = ", skm.mean_absolute_error(y_test, predicted))
```

```python
print("Test     Predicted")
for i, j in zip(y_test, predicted):
    print(i, " ", j)
```
