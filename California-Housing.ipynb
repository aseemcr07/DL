```python
from sklearn.datasets import fetch_california_housing
import pandas as pd
```

```python
california = fetch_california_housing()
```

```python
california.data.shape
```

```python
california.target.shape
```

```python
california.feature_names
```

```python
pd.set_option('precision', 4)
pd.set_option('max_columns', 9)
pd.set_option('display.width', None)
california_df = pd.DataFrame(california.data,
                             columns=california.feature_names)
california_df['MedHouseValue'] = pd.Series(california.target)
california_df.head()
```

```python
california_df.shape
```

```python
data_col = california_df.columns
```

```python
for col in data_col:
    print(col)
```

```python
pred = california_df[data_col[data_col!="MedHouseValue"]]
targ = california_df["MedHouseValue"]
```

```python
pred
```

```python
targ
```

```python
#data splitting into training and testing
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(pred, targ, random_state=104, test_size=0.25, shuffle=True)
```

```python
n_cols = pred.shape[1]
n_cols
```

```python
import keras
```

```python
from keras.models import Sequential
from keras.layers import Dense
```

```python
model = Sequential()
```

```python
model.add(Dense(128, activation="relu", input_shape = (n_cols, )))
model.add(Dense(50, activation="relu"))
model.add(Dense(50, activation="relu"))
model.add(Dense(1))

model.summary()
```

```python
model.compile(optimizer = "rmsprop", loss="mse", metrics=['mae'])
```

```python
# Train the model
model.fit(X_train, y_train, validation_split=0.3, epochs=200, verbose=1)
```

```python
import sklearn.metrics as skm
```

```python
predicted = model.predict(X_test)
```

```python
print("MAE = ", skm.mean_absolute_error(y_test, predicted))
```

```python
for i,j in zip(y_test, predicted):
    print(i, " ", j)
```
