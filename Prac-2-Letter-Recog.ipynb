```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import preprocessing
```

```python
cols =["res","x1","x2","x3","x4","x5","x6","x7","x8","x9","x10","x12","x13","x14","x15","x16","x17"]
data = pd.read_csv("letter-recognition.data", sep=',', names=cols)
data.head()
```

```python
label_encoder = preprocessing.LabelEncoder()
data_col = data.columns
data["res"] = label_encoder.fit_transform(data["res"])
lable = data["res"].unique()
pred = data[data_col[data_col!="res"]]
tar = data["res"]
print(lable)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test,y_train, y_test = train_test_split(pred,tar, random_state=104, test_size=0.25, shuffle=True)
```

```python
print(X_test)
```

```python
print(y_test)
```

```python
n_cols = pred.shape[1]
print(n_cols)
```

```python
import keras
from keras.models import Sequential
from keras.layers import Dense
```

```python
model = Sequential()
model.add(Dense(128, activation= "relu", input_shape = (n_cols,)))
model.add(Dense(50,activation = "relu"))
model.add(Dense(50,activation = "relu"))
model.add(Dense(26,activation = "softmax"))
model.compile(optimizer="adam", loss="sparse_categorical_crossentropy", metrics=["accuracy"])

model.summary()

model.fit(X_train,y_train,validation_split = 0.3, epochs = 20, verbose =1)
```

```python
predicted = model.predict(X_test)

valu = []

for i in predicted:
    valu.append(np.argmax(i,0))

import sklearn.metrics as metric
print("Mean absolute error =", round(metric.mean_absolute_error(y_test, valu), 2))

for i,j in zip(y_test,valu):
    print(i," ",j)
```
